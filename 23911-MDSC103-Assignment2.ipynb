#**Creating Gradient Desent for Dynamic functions in 2D**

This is the gradient function where it takes 7 parameteres


*   f - Function
*   df - Gradient Function
*   neta - Learning rate 
*   x0 - Initial value for iteration
*   tol - tolerance()
*   ran - Intervel in which we are looking
*   iter - Value of Terminating iteration





---

***Note:This function will either print the x value or it will state that it went out of iterations.***

---



```python
def gradient(f,df,neta,x0,tol,ran,iter):
    Xpoints=[x0]
    Ypoints=[f(x0)]

    for _ in range(iter):

        x1=x0-(neta*df(x0))

        Xpoints.append(x0)
        Ypoints.append(f(x0))

        if abs(x1-x0)<tol:
            print("x\t\t:",x1,"\nf(x)\t:",f(x1),end="\n\n")
            break

        x0=x1

    else:
        print('\nOut of Iterrations\n')

    X=np.linspace(ran[0],ran[1],100)
    plt.plot(X,f(X))

    plt.xlabel('X Values')
    plt.ylabel('Function Values')
    plt.title('Gradient Decent')

    plt.plot(Xpoints,Ypoints,color='r',marker='o')
    plt.grid()
    plt.show()
```
