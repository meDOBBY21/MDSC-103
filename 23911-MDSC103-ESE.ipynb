```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x0=[-1,-1]
F = lambda X:-(4*X[0]+6*X[1]-2*X[0]**2-2*X[0]*X[1]-2*X[1]**2)
sp.optimize.minimize(F,x0)
```

```python
x1=np.linspace(-10,10,100)
x2=np.linspace(-9,11,100)

f = lambda x1,x2:-(4*x1+6*x2-2*x1**2-2*x1*x2-2*x2**2)

X1,X2=np.meshgrid(x1,x2)
y=f(X1,X2)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X1, X2, y,cmap='jet')
plt.show()
```

#Gradient Decent

```python
x0=np.array([2,2])
lr=0.1
dj1=lambda X:-4+4*X[0]+2*X[1]
dj2=lambda X:-6+2*X[0]+4*X[1]
for i in range(100):

    vector = np.array([dj1(x0),dj2(x0)])

    x1=x0-(lr*vector)

    x0=x1
print(f'Final Solution is x1:{x1[0]} and x2:{x1[1]}')
```

#Plotting the minimum point

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X1, X2, y,cmap='jet')
ax.plot(x1[0],x1[1],F(x1),marker='^',color='red')
plt.show()
```
