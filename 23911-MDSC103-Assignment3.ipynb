A_eq @ x == b_eq

```python
from scipy.optimize import linprog
import matplotlib.pyplot as plt
import numpy as np
```

Maximize:

$$z=2x1+3x2$$
subj:

$$2x1+x2<=4$$
$$x1+2x2<=5$$
$$x1,x2>=0$$




```python
c=[-2,-3]
A=[[2,1],[1,2]]
b=[4,5]
x0_bound=(0,None)
x1_bound=(0,None)

res=linprog(c,A_ub=A,b_ub=b,bounds=[x0_bound,x1_bound])

res
```

```python
c1=lambda x:4-(2*x)
c2=lambda x:(5-x)/2

x=np.linspace(-2,5,10)
y1=c1(x)
y2=c2(x)

plt.axhline(y=0)
plt.axvline(x=0)

plt.plot(x,y1,color='r',alpha=.5,label='4-2x')
plt.plot(x,y2,color='g',alpha=.5,label='(5-x)/2')


x1=np.linspace(1,2,10)
x2=np.linspace(0,1,10)

plt.fill_between(x1,c1(x1),color='yellow',alpha=.5)
plt.fill_between(x2,c2(x2),color='yellow',alpha=.5)

plt.grid()
plt.legend()
plt.show()
```
